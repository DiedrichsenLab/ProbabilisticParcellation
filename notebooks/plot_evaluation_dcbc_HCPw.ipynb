# DCBC evaluation

```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import sys

base_dir = '/Volumes/diedrichsen_data$/data/FunctionalFusion'
if not Path(base_dir).exists():
    base_dir = '/srv/diedrichsen/data/FunctionalFusion'
```

This notebook plots the DCBC and cosine error evaluation of the different parcellations

## Evaluating different HCP weights

```python

evaltasks = f'{base_dir}/Models/Evaluation_04/eval_dcbc_sym_K-34_task_34.tsv'
evalhcp = f'{base_dir}/Models/Evaluation_04/eval_dcbc_sym_K-34_HCPw_sym.tsv'
D1 = pd.read_csv(evaltasks, sep = '\t')
D2 = pd.read_csv(evalhcp, sep='\t')
D = pd.concat([D1, D2], ignore_index=True)

```

```python
plt.figure(figsize=(15,8))
sb.barplot(data=D,x='test_data',y='dcbc_group', hue='model_name')
plt.ylim(0, 0.13)
```
